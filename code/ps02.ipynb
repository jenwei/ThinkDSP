###### 1) Run and listen to the examples in chap03.ipynb. Replace the Hamming window in the leakage example with one of the other windows provided by Numpy and see what effect they have on leakage.

Using the scipy numpy reference, I chose to compare the Hamming window with the Bartlett window.
http://docs.scipy.org/doc/numpy/reference/routines.window.html

The Hamming window, not to be confused with the Hanning window, is a taper formed by using a weighted cosine, while the Bartlett window is essentially a triangular window with end points at zero.

```python
signal = thinkdsp.SinSignal(freq=440)
duration = signal.period * 30.5
wave0 = signal.make_wave(duration)
wave0.plot()
thinkplot.config(xlabel='time (s)', ylabel='amplitude', legend=False)
```

Above is code taken from chap03 where the signal we're working with is a Sine signal with leakage due to the non-multiple for the duration.

```python
spectrum = wave0.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Above is the original spectrum plot of the Sine signal.

```python
wave0.hamming()
spectrum = wave0.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Above is the spectrum plot of the Sine signal with the Hamming window. When compared to the original plot, notice how the ends drop down to zero earlier, meaning that there is less leakage.

Now, going back to the original Sine signal, a different window, the Bartlett window, is applied.

```python
wave1 = signal.make_wave(duration)
wave1.ys *= numpy.bartlett(len(wave1.ys))
spectrum = wave1.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Above is the spectrum plot of the Sine signal with the Bartlett window.

Compared to the original Sine signal spectrum plot, both the Hamming and Bartlett windows reduce the leakage. When the two windows are compared against each other, in the plots above, the Hamming seems a tad more effective as it has a steeper dropoff, so the leakage seems to only spread as far as 400 Hz and 475 Hz as opposed to that of the Bartlett window which leaks from about 350 Hz to 525 Hz. In general, the purpose of windows is to reduce leakage by reducing the amplitudes of a signal in such a way that, in a sense, hides the fact that the signal is not periodic.

###### 2a) Write a SawtoothChirp class that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decrease) linearly).

Hint: [Combine the evaluate functions from Chirp and SawtoothSignal]

```python
import math

class SawtoothChirp(thinkdsp.Chirp):
    """Represents a sawtooth signal with varying frequency."""
    
    def _evaluate(self, ts, freqs):
        """Function to evaluate signal
        ts: float array of times
        freqs: float array of requencies during each interval"""
        dts = numpy.diff(ts)
        dps = (2*math.pi) * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        #Code above this point is the same as Chirp#
        cycles = phases / (2*math.pi) #Given - no need to recalculate as done in Sawtooth
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys
        #Code above this point is the same (more or less) as the Sawtooth#
```

As suggested/implied by the hint, the Sawtooth Chirp is a combination of the Chirp and SawtoothSignal evaluate functions. Comments in the code separate the evaluate function into its sections of Chirp code and Sawtooth code. The SawtoothChirp evaluate function works by taking the phases from Chirp and then calculating out the ys using that and plugging it into the ys equation from Sawtooth.

###### 3) In musical terminology, a "glissando" is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to the length.

###### Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

###### Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3 (C3 = 262 Hz, F3 = 349 Hz).

```python
low = thinkdsp.ExpoChirp(start=262, end=349)
high = thinkdsp.ExpoChirp(start=349, end=262)
low_wave = low.make_wave(duration=2)
high_wave = high.make_wave(duration=2)
low_wave.apodize()
high_wave.apodize()
gliss=low_wave|high_wave
gliss.make_audio()
```

```python
def tromboneGliss(start = 100, end = 100, duration = 2):
    """tromboneGliss takes in two inputs, 
    a start and end frequency, and creates 
    a glissando that slides at a constant 
    speed"""
    low = thinkdsp.ExpoChirp(start, end)
    high = thinkdsp.ExpoChirp(start, end)
    low_wave = low.make_wave(duration)
    high_wave = high.make_wave(duration)
    low_wave.apodize()
    high_wave.apodize()
    gliss=low_wave|high_wave
    return gliss

TGliss = tromboneGliss(start=262, end=349, duration=5)
TGliss.make_audio()
```

Above is code that simulates a trombone glissando from C3 up to F3 and back down. The trombone glissando uses ExpoChirp instead of Chirp to recreate the steady, constant speed since we ant pitch to increase linearly. Thus, the frequency must increase exponentially.

###### 6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random. To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

###### Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

For this question, I will continue working with the wind audio file.

```python
wind = wind_blowing.segment(start=2.0, duration=5.0)
```

```python
def bartlett(wave, m=2):
    """
    bartlett takes in a wave and segment length, m (in seconds)
    
    bartlett then returns the average power array and
    its corresponding frequencies
    """
    powers = 0
    freq = 0
    segs = numpy.floor_divide(wave.duration,m) #total number of possible segments
    for i in range(0,int(m*segs),m):
        s = wave.segment(i, m) #extracting a subsegment of the wave segment
        spec = s.make_spectrum() #make a spectrum
        p = spec.power #get the power
        powers = powers + p
        freq = freq + spec.fs
    avg = powers/segs
    return avg, freq
```

```python
avg_pow, frequency = bartlett(wind,1)
thinkplot.Plot(frequency, avg_pow)
thinkplot.config(xlabel='frequency (Hz)', ylabel='power', xscale='log', yscale='log')
```

Above is the resulting plot from using the Bartlett Method and below is the original power spectrum plot.

```python
wind.make_spectrum().plot_power(label='Wind')
thinkplot.show(xlabel='freq', ylabel='power density', xscale='log', yscale='log',legend=True)
```

The Bartlett Method seems to smooth and refined, giving a clearer picture of the relationship between power and frequency.
